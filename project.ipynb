```python
#libraries and data imports

import numpy as np # linear algebra

import pandas as pd # data processing, CSV file I/O 

from matplotlib import pyplot as plt

import seaborn as sns

from sklearn.preprocessing import LabelEncoder

import copy

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

airlinedata = pd.read_csv("Predict Price of Airline Tickets.csv")

airlinedata#this will be our main data
```

```python
#this will be our data used for the machine learing

airlinedata_encoded = copy.deepcopy(airlinedata)

airlinedata_encoded
```

```python
airlinedata_encoded.isnull().sum()#finding the how many null values we have
```

```python
airlinedata_encoded.dropna(inplace= True)#as they are low in number we drop thenm

airlinedata_encoded.isnull().sum()
```

```python
#we will seperate the date of journey to 2 integer values in day and month to ease the processing

airlinedata_encoded["Day_of_Journey"]=airlinedata_encoded.Date_of_Journey.str.split("/").str[0].astype(int)

airlinedata_encoded["Month_of_Journey"]=airlinedata_encoded.Date_of_Journey.str.split("/").str[1].astype(int)

#we drop the coulmns we replaced 

airlinedata_encoded.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
airlinedata_encoded
```

```python
#like we did with the date we will do with the depature time and arrival time but to hours and mintes

airlinedata_encoded["Departure_Hour"]=0

airlinedata_encoded["Departure_Min"]=0

airlinedata_encoded["Arrival_Hour"]=0

airlinedata_encoded["Arrival_Min"]=0

#we made a coulmn for each with the above lines so we can access them with indexes in the for loop

for i in range(9039):#will raise error at 9039 because it was removed when we droped the nulls
    
    airlinedata_encoded.at[i,"Departure_Hour"]=int(airlinedata_encoded.at[i,"Dep_Time"].split(":")[0])
    
    airlinedata_encoded.at[i,"Departure_Min"]=int(airlinedata_encoded.at[i,"Dep_Time"].split(":")[1])
    
    
    #the last split in the arrival mins to remove the unneeded date
    
    
for i in range(9040,len(airlinedata_encoded)+1):#so we stop before 9039 and continue after it
    
    airlinedata_encoded.at[i,"Departure_Hour"]=int(airlinedata_encoded.at[i,"Dep_Time"].split(":")[0])
    
    airlinedata_encoded.at[i,"Departure_Min"]=int(airlinedata_encoded.at[i,"Dep_Time"].split(":")[1])
    
    
airlinedata_encoded["Arrival_hour"] = pd.to_datetime(airlinedata_encoded.Arrival_Time).dt.hour

airlinedata_encoded["Arrival_min"] = pd.to_datetime(airlinedata_encoded.Arrival_Time).dt.minute


airlinedata_encoded
```

```python
#we drop the coulmns we replaced 

airlinedata_encoded.drop(["Dep_Time"], axis = 1, inplace = True)

airlinedata_encoded.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
airlinedata_encoded.Total_Stops.unique()#we have a look at the unique values of the total stops
```

```python
#we will use the numbers already listed in it by seperating the int from the string and if it is a non stop we replace with 0

for i in range(9039):#will raise error at 9039 because it was removed when we droped the nulls
    
    if airlinedata_encoded.Total_Stops.at[i]=='non-stop':
        
        airlinedata_encoded.Total_Stops.at[i]=0
        
    else:
        
        airlinedata_encoded.Total_Stops.at[i]=int(airlinedata_encoded.Total_Stops.at[i].split()[0])
        

for i in range(9040,len(airlinedata_encoded)+1):
    
    if airlinedata_encoded.Total_Stops.at[i]=='non-stop':
        
        airlinedata_encoded.Total_Stops.at[i]=0
        
    
    else:
        
        airlinedata_encoded.Total_Stops.at[i]=int(airlinedata_encoded.Total_Stops.at[i].split()[0])
        
```

```python
airlinedata_encoded.Total_Stops=pd.to_numeric(airlinedata_encoded.Total_Stops)#to change the data type to int

airlinedata_encoded.Total_Stops
```

```python
# here we seperate the route coulmn to 5 route points which is the maximum number of routes

airlinedata_encoded['Route1']=airlinedata_encoded['Route'].str.split('?').str[0]#we use the split and split with the mark between them

airlinedata_encoded['Route2']=airlinedata_encoded['Route'].str.split('?').str[1]

airlinedata_encoded['Route3']=airlinedata_encoded['Route'].str.split('?').str[2]

airlinedata_encoded['Route4']=airlinedata_encoded['Route'].str.split('?').str[3]

airlinedata_encoded['Route5']=airlinedata_encoded['Route'].str.split('?').str[4]

for i in ['Route1', 'Route2', 'Route3', 'Route4', 'Route5']:#and then we encode them from strings to numbers
    
    airlinedata_encoded[i]=LabelEncoder().fit_transform(airlinedata_encoded[i])

airlinedata_encoded
```

```python
#we drop the coulmn as we don't need route because we mainly need the number of stops and destnation and source from the other coulmns

#and the duration can be transformed to number of mins

airlinedata_encoded.drop(["Route"], axis = 1, inplace = True)

airlinedata_encoded["Duration"]=airlinedata_encoded['Duration'].str.replace("h", '*60').str.replace(' ','+').str.replace('m','*1').apply(eval)
```

```python
airlinedata_encoded
```

```python
airlinedata_encoded.Additional_Info.unique()#checking the additional info uniques
```

```python
#in the additional info there exists same value but just different case sensitfty so we equalize them using the lower function

for i in range(9039):
    
    airlinedata_encoded.Additional_Info.at[i]=airlinedata_encoded.Additional_Info.at[i].lower()
        

for i in range(9040,len(airlinedata_encoded)+1):
    
    airlinedata_encoded.Additional_Info.at[i]=airlinedata_encoded.Additional_Info.at[i].lower()
```

```python
airlinedata_encoded.Additional_Info.unique()
```

```python
#here we use the one-hot encoding method to encode the Airline,Source,Destination,Additional_Info coulmns

Airline=pd.get_dummies(airlinedata_encoded[['Airline','Source','Destination','Additional_Info']]) 

airlinedata_encoded=pd.concat([airlinedata_encoded, Airline],axis=1)

#and then drop them

airlinedata_encoded.drop(['Airline','Source','Destination','Additional_Info'], axis = 1, inplace = True)
```

```python

airlinedata_encoded
```

```python
#here we will remove the outliers of the price and duration using the iqr method

for x in ["Price","Duration"]:
    
    q75,q25 = np.percentile(airlinedata_encoded.loc[:,x],[75,25])
    
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    
    min = q25-(1.5*intr_qr)
 
    airlinedata_encoded.loc[airlinedata_encoded[x] < min,x] = np.nan#and here we replace them with nans
    
    airlinedata_encoded.loc[airlinedata_encoded[x] > max,x] = np.nan

airlinedata_encoded.dropna(inplace= True)#and drop them with the drop na method
```

```python
#now we seperate our x and y data
airlinedata_encoded.sort_values(by=['Price'],inplace=True)

y = airlinedata_encoded["Price"]

X = airlinedata_encoded.drop(columns='Price',axis=1)

X
```

```python
#and then seperate them to train and test 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/4, random_state = 0)

X_train
```

```python
#now we make our linearregression training model

regressor = LinearRegression()

regressor.fit(X_train, y_train)
```

```python
#here we predict the test resaults

y_pred = regressor.predict(X_test)

y_pred
```

```python
#this is our model's score

print('Training Score :',regressor.score(X_train, y_train)*100,'%')

print('Test Score     :',regressor.score(X_test, y_test)*100,'%')
```
